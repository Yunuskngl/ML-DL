```python
import matplotlib.pyplot as plt
import numpy as np

def generate_visualization(terminal_left_reward, terminal_right_reward, each_step_reward, gamma, misstep_prob):
    num_states = 6
    num_actions = 2
    
    # Q(s, a) değerlerini hesaplayalım
    Q = np.zeros((num_states, num_actions))
    
    # Optimal politika ve ödülleri hesaplayalım
    optimal_policy = []
    rewards = []
    
    for state in range(num_states):
        left_value = terminal_left_reward * (gamma ** state)
        right_value = terminal_right_reward * (gamma ** (num_states - 1 - state))
        
        if left_value > right_value:
            optimal_policy.append('←')
            rewards.append(left_value)
            Q[state, 0] = left_value
        else:
            optimal_policy.append('→')
            rewards.append(right_value)
            Q[state, 1] = right_value
            
    # Görselleştirme kısmı
    fig, axs = plt.subplots(2, 1, figsize=(10, 6))
    
    axs[0].set_title("Optimal policy")
    axs[0].table(cellText=[rewards, optimal_policy], loc='center', cellLoc='center')
    axs[0].axis('off')
    
    axs[1].set_title("Q(s, a)")
    Q_table = np.round(Q, decimals=2).tolist()
    axs[1].table(cellText=Q_table, loc='center', cellLoc='center')
    
    axs[1].axis('off')
    
    plt.show()



```

```python

num_states = 6
num_actions = 2

terminal_left_reward = 100
terminal_right_reward = 40
each_step_reward =0

#discount factor
gamma = 0.5

#probability of going in the wrong direction
misstep_prob = 0

generate_visualization(terminal_left_reward, terminal_right_reward, each_step_reward, gamma, misstep_prob)
```
